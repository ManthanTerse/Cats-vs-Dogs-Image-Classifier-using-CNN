# CNN-based Cats vs Dogs Image Classification

```python
import tensorflow as tf
from keras.layers import Dense , Conv2D , MaxPooling2D, Flatten, Input, BatchNormalization, Dropout
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Dataset used from kaggle:

```python
import kagglehub

path = kagglehub.dataset_download("salader/dogsvscats")

print("Path to dataset files:", path)
```

```python
#CNN Architecture:
cnn = Sequential()
cnn.add(Input(shape=(64,64,3)))

cnn.add(Conv2D(filters=32, kernel_size=(3,3), input_shape=(64,64,3), activation='relu'))
cnn.add(BatchNormalization()) 
cnn.add(MaxPooling2D(pool_size=(2,2)))

cnn.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=(2,2)))

cnn.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=(2,2)))

cnn.add(Flatten())

cnn.add(Dense(units=128, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(Dropout(0.5))

cnn.add(Dense(units=64, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(Dropout(0.3))

cnn.add(Dense(1,activation='sigmoid'))
```

```python
cnn.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3)
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

training_set = train_datagen.flow_from_directory(
    r'C:\Users\manthan\.cache\kagglehub\datasets\salader\dogsvscats\versions\1\train', 
    target_size=(64,64), 
    batch_size=32, 
    class_mode='binary'
)
test_set = test_datagen.flow_from_directory(
    r'C:\Users\manthan\.cache\kagglehub\datasets\salader\dogsvscats\versions\1\test', 
    target_size=(64,64), 
    batch_size=32, 
    class_mode='binary'
)

cnn.fit(training_set, epochs=50, validation_data=test_set, callbacks=[early_stop])
```

```python
cnn.save("cats_dogs_model.keras")
```

```python
from tensorflow.keras.models import load_model

cnn = load_model("cats_dogs_model.keras")
```

```python
import numpy as np
from keras.preprocessing import image

test_image = image.load_img(r"C:\Users\manthan\Python\DL\Projects\Images\7.jpg", target_size=(64,64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0) 
result = cnn.predict(test_image)
if result > 0.5:
    print("Prediction: Cat")
else:
    print("Prediction: Dog")
```
